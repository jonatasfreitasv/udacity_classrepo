```python
import pandas as pd

countries = ['Albania', 'Algeria', 'Andorra', 'Angola', 'Antigua and Barbuda',
             'Argentina', 'Armenia', 'Australia', 'Austria', 'Azerbaijan',
             'Bahamas', 'Bahrain', 'Bangladesh', 'Barbados', 'Belarus',
             'Belgium', 'Belize', 'Benin', 'Bhutan', 'Bolivia']

life_expectancy_values = [74.7,  75. ,  83.4,  57.6,  74.6,  75.4,  72.3,  81.5,  80.2,
                          70.3,  72.1,  76.4,  68.1,  75.2,  69.8,  79.4,  70.8,  62.7,
                          67.3,  70.6]

gdp_values = [ 1681.61390973,   2155.48523109,  21495.80508273,    562.98768478,
              13495.1274663 ,   9388.68852258,   1424.19056199,  24765.54890176,
              27036.48733192,   1945.63754911,  21721.61840978,  13373.21993972,
                483.97086804,   9783.98417323,   2253.46411147,  25034.66692293,
               3680.91642923,    366.04496652,   1175.92638695,   1132.21387981]

le = pd.Series(life_expectancy_values, index=countries)
gdp = pd.Series(gdp_values, index=countries)
```

```python
# Correlation data

both_above = (le > le.mean()) & (gdp > gdp.mean())
both_below = (le < le.mean()) & (gdp < gdp.mean())

is_same_direction = (both_above | both_below).sum()

is_not_same_direction = len(gdp) - is_same_direction

print('In same direction = %s' % is_same_direction)
print('Not same direction = %s' % is_not_same_direction)

if is_same_direction > is_not_same_direction:
    print('Positive Correlation in this data')
if is_same_direction < is_not_same_direction:
    print('Negative Correlation in this data')
if (is_same_direction - is_not_same_direction) < 10: #attention, this is a magic number
    print('Probably not have any correlation in this data')
```

```python
# Max values
print('Country with max life expectancy is %s with age %s' % (le.argmax(), le.loc[le.argmax()]))
print('Country with max gdp is %s with value %s' % (gdp.argmax(), gdp.loc[gdp.argmax()]))
```
